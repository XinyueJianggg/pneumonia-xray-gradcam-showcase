# Pneumonia X-ray Classification with Grad-CAM

This is a public showcase notebook for a chest X-ray pneumonia classification project. It contains only selected summaries and output figures. The full course notebook and implementation are kept private.


## Project Goal

The goal was to compare several CNN-based models for binary pneumonia classification from chest X-ray images. The project focused on both prediction quality and interpretability.

Because this is a medical screening task, the most important evaluation question was not only “how accurate is the model?” but also “how many pneumonia cases does the model miss?”


## Models Compared

The full private notebook compared:

- Custom CNN baseline
- ResNet-like residual CNN
- ResNet18-style CNN
- DenseNet121 with ImageNet transfer learning
- A 224×224 DenseNet121 experiment for clearer Grad-CAM visualization

DenseNet121 was the strongest model overall. The ResNet18-style model reached high training performance but generalized poorly, suggesting overfitting or training instability.


## DenseNet121 Test Metrics

The table below reports manually computed pneumonia-class metrics for DenseNet121. These are more clinically meaningful than the raw built-in softmax precision/recall metrics.

| Metric | Value |
|---|---:|
| Accuracy | 0.8990 |
| Precision | 0.8691 |
| Recall / Sensitivity | 0.9872 |
| Specificity | 0.7521 |
| False Negative Rate | 0.0128 |
| ROC-AUC | 0.9812 |
| True Negative | 176 |
| False Positive | 58 |
| False Negative | 5 |
| True Positive | 385 |

The model missed 5 pneumonia cases in the test set. This gives a high pneumonia recall of 98.72%, which is useful for a screening-oriented model. The tradeoff is lower specificity, meaning more normal cases were flagged as pneumonia.

## Grad-CAM Interpretability

Grad-CAM was used as a qualitative interpretability tool. It does not prove that the model uses correct clinical reasoning, but it helps check whether the model appears to focus on meaningful image regions.

The 128×128 DenseNet model performed strongly numerically, while the 224×224 experiment produced clearer heatmaps because the final convolutional feature map had higher spatial resolution.

## Key Takeaways

1. DenseNet121 with transfer learning gave the strongest overall generalization.
2. Manual pneumonia-class metrics are more useful than generic softmax metrics for this medical screening problem.
3. High recall and low false negative rate matter because false negatives are missed pneumonia cases.
4. Grad-CAM provides a qualitative check of model focus, but it should not be interpreted as clinical proof.
5. Higher-resolution inputs can improve Grad-CAM clarity even when the numerical metrics do not improve.
